# Inspecting real-time data snapshots
There have been noticeable cases where real-time requests on journeys will have heavy delays, then future requests will have reset these lines to zero.
93-A.1260.122.36.A.8.55187054

```python
import glob
from google.transit import gtfs_realtime_pb2
import pickle

data_path = 'home/pi/sydney-transport-tracker/data/raw/20190124/'
delay_files = sorted(glob.glob(data_path + '*.pickle'))

for delay_data_file in delay_files:
    current_delay_response = pickle.load(open(delay_data_file, "rb"))

    feed = gtfs_realtime_pb2.FeedMessage()
    feed.ParseFromString(current_delay_response)
    for entity in feed.entity:
        
        if entity.trip_update.trip.trip_id == "93-A.1260.122.36.A.8.55187054":
            print(delay_data_file)
            #print(entity.trip_update.trip.schedule_relationship)
            print(entity)
            #for stop_time_update in entity.trip_update.stop_time_update:
             #   if stop_time_update.stop_id == '204261':
              #      print(stop_time_update)
                    
```
